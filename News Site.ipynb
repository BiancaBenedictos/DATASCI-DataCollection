# Getting News Articles Published on March 11-12

For this notebook, NewsAPI was used to extract news articles published by ABC News (Australian Broadcasting Corporation)

### Import libraries

```python
from newsapi import NewsApiClient
from pandas.io.json import json_normalize
import pandas as pd
import json
```

### Initialize API key 
It would be better if one makes his/her api key private or not visible in the notebook

### Get all news articles by ABC News AU
Modify parameters as needed. 

```python
# /v2/everything
all_articles = newsapi.get_everything(q=' ',
                                      sources='abc-news-au',
                                      domains='http://www.abc.net.au/news',
                                      from_param='2021-03-11',
                                      to='2021-03-12',
                                      language='en',
                                      sort_by='relevancy',
                                      page=2)
```

```python
all_articles
```

### Create Data
Flatten the articles attribute which contains all information of the articles retrieved via NewsAPI

```python
news = json_normalize(all_articles, 'articles')
```

Drop unnecessary columns and rename the remaining columns

```python
news.drop(['urlToImage', 'description', 'source.id', 'source.name'],
  axis='columns', inplace=True)
news
```

```python
news.rename(columns = {'publishedAt': 'date', 'content': 'full_article'}, inplace = True)
news
```

### Create JSON file
Convert dataframe to json and generate json file

```python
result = news.to_json(orient="records")
parsed = json.loads(result)
json.dumps(parsed, indent=4)  
```

```python
with open("abc-news-au_data", 'w') as outfile:
    json.dump(parsed, outfile)
```
